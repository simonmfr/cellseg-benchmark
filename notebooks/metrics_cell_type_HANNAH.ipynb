## Compute cell type distribution

````python
from cellseg_benchmark.metrics import (
    compute_cell_type_distribution,
    compute_metric_for_all_methods,
)

compute_metric_for_all_methods(
    compute_cell_type_distribution,
    results_name="cell_type_metrics/cell_type_distribution_adata_integrated_cell_type_revised.csv",
    cohort="aging",
    celltype_name="cell_type_revised",
    adata_name="adata_integrated",
    overwrite=False,
)

compute_metric_for_all_methods(
    compute_cell_type_distribution,
    results_name="cell_type_metrics/cell_type_distribution_adata_vascular_subset_cell_type_incl_zonation.csv",
    cohort="aging",
    celltype_name="cell_type_incl_zonation",
    adata_name="adata_vascular_subset",
    overwrite=False,
)
````

## Plot cell type distribution

````python
from cellseg_benchmark.metrics import plot_cell_type_distribution

plot_cell_type_distribution("aging", "adata_integrated_cell_type_revised", show=True)
plot_cell_type_distribution(
    "aging", "adata_vascular_subset_cell_type_incl_zonation", show=True
)
````

```python
#check if adata contains leiden clustering. It does not -> needs to be computed. 
from cellseg_benchmark import BASE_PATH
from pathlib import Path
import scanpy as sc
cohort = 'aging'
method = 'Cellpose_1_nuclei_model'
adata_name = 'adata_integrated'
data_path = Path(BASE_PATH) / "analysis" / cohort / method
adata_path = data_path / "adatas" / f"{adata_name}.h5ad.gz"
adata = sc.read_h5ad(adata_path)
adata.obs.columns
```

````python
from cellseg_benchmark.metrics import (
    compute_clustering_scores,
    compute_metric_for_all_methods,
)

compute_metric_for_all_methods(
    compute_clustering_scores,
    results_name="cell_type_metrics/clustering_score_adata_integrated_leiden.csv",
    cohort="aging",
    celltype_name="leiden",
    adata_name="adata_integrated",
    overwrite=False,
)

compute_metric_for_all_methods(
    compute_clustering_scores,
    results_name="cell_type_metrics/clustering_score_adata_vascular_subset_leiden.csv",
    cohort="aging",
    celltype_name="leiden",
    adata_name="adata_vascular_subset",
    overwrite=False,
)
````

## Plot clustering scores

````python
from cellseg_benchmark.metrics import plot_clustering_scores

plot_clustering_scores("aging", "adata_integrated_leiden", show=True)
plot_clustering_scores("aging", "adata_vascular_subset_leiden", show=True)
````
