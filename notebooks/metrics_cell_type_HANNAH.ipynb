## Compute cell type distribution

```python
from cellseg_benchmark.metrics import (
    compute_cell_type_distribution,
    compute_metric_for_all_methods,
)

compute_metric_for_all_methods(
    compute_cell_type_distribution,
    cohort="aging",
    celltype_name="cell_type_revised",
    adata_name="adata_integrated",
    overwrite=False,
)

compute_metric_for_all_methods(
    compute_cell_type_distribution,
    cohort="aging",
    celltype_name="cell_type_incl_zonation",
    adata_name="adata_vascular_subset",
    overwrite=False,
)
```

## Plot cell type distribution

```python
from cellseg_benchmark.metrics import plot_cell_type_distribution

plot_cell_type_distribution("aging", "adata_integrated_cell_type_revised", show=True)
plot_cell_type_distribution(
    "aging", "adata_vascular_subset_cell_type_incl_zonation", show=True
)
```

## Compute clustering scores

```python
from cellseg_benchmark.metrics import (
    compute_clustering_scores,
    compute_metric_for_all_methods,
)

compute_metric_for_all_methods(
    compute_clustering_scores,
    cohort="aging",
    celltype_name="leiden",
    adata_name="adata_vascular_subset",
    overwrite=False,
)
```

## Plot clustering scores

```python
from cellseg_benchmark.metrics import plot_clustering_scores

plot_clustering_scores("aging", "adata_integrated_leiden", show=True)
plot_clustering_scores("aging", "adata_vascular_subset_leiden", show=True)
```
