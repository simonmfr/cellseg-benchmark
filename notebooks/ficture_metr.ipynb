```python
import pandas as pd
import seaborn as sns
from anndata import read_h5ad
```

```python
adata_1 = read_h5ad("/Users/jonasflor/Downloads/adata.h5ad")
data_mean_1 = pd.concat(
    (
        adata_1.obs[["cell_type_mmc_raw_revised", "full_name"]],
        adata_1.obsm["ficture21_mean"],
    ),
    axis=1,
)
means_1 = data_mean_1.copy()
del means_1["full_name"]
means_1 = means_1.groupby("cell_type_mmc_raw_revised").mean()
fac_2 = data_mean_1[
    ["cell_type_mmc_raw_revised", "full_name", "fictureF21_2_mean_intensity"]
]
fac_2_EC = fac_2[fac_2["cell_type_mmc_raw_revised"] == "ECs"]
fac_2_SMC = fac_2[fac_2["cell_type_mmc_raw_revised"] == "SMCs"]
fac_2_Peri = fac_2[fac_2["cell_type_mmc_raw_revised"] == "Pericytes"]
```

```python
sns.histplot(
    fac_2_EC, x="fictureF21_2_mean_intensity", hue="full_name", multiple="stack"
)
```

```python
sns.histplot(
    fac_2_SMC, x="fictureF21_2_mean_intensity", hue="full_name", multiple="stack"
)
```

```python
sns.histplot(
    fac_2_Peri, x="fictureF21_2_mean_intensity", hue="full_name", multiple="stack"
)
```
